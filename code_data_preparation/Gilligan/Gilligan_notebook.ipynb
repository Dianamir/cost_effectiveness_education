```python
from scipy import stats

print(stats.ks_2samp(
    df[df.treatment==0].pre_theta,
    df[df.treatment==1].pre_theta
))
stats.ttest_ind(
    df[df.treatment==0].pre_theta,
    df[df.treatment==1].pre_theta,
    nan_policy='omit'
)
```

```python
ks_list = []
ttest_list = []
for dec in list(df.decile.unique()):
    df_tmp = df[df.decile==dec]
    print(f'decile number :   {dec}', '\n\n')
    ks = stats.ks_2samp(
        df_tmp[df_tmp.treatment==0].pre_pp_b04,
        df_tmp[df_tmp.treatment==1].pre_pp_b04
    )
    ks_list.append(ks.pvalue)
    print(ks)
    print('\n')
    ttest = stats.ttest_ind(
        df_tmp[df_tmp.treatment==0].pre_pp_b04,
        df_tmp[df_tmp.treatment==1].pre_pp_b04,
        nan_policy='omit'
    )
    ttest_list.append(ttest.pvalue)
    print(ttest)
    print('\n')
    print('_'*30)
```

```python
np.mean(ks_list), np.mean(ttest_list)
```
